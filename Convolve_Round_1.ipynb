```python
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

### Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

### Logistic Regression Model without handling imbalanced classes

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X_normalized,y, test_size=0.2, random_state=42)

lr_1 = LogisticRegression()
lr_1.fit(X_train, y_train)
y_pred_1 = lr_1.predict(X_test)

print(classification_report(y_test,y_pred_1))
```

### Logistic Regression Model with Handling Imbalanced Classes

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=1)
X_train_resampled, y_train_resampled = sm.fit_resample(X_train, y_train)
```

```python
plt.bar(pd.Series(y_train_resampled).value_counts().index, pd.Series(y_train_resampled).value_counts().values, color='orange', alpha=0.7, label='After SMOTE')

plt.legend()
plt.show()
```

```python
lr = LogisticRegression()
lr.fit(X_train_resampled, y_train_resampled)
y_pred = lr.predict(X_test)

print(classification_report(y_test,y_pred))
```
